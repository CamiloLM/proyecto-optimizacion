## FORMULACIÓN MATEMÁTICA COMPLETA

### Conjuntos y Parámetros

| Elemento | Notación Matemática | Descripción |
| :--- | :--- | :--- |
| **Filas** | $I = \{1, 2, \dots, m\}$ | Conjunto de elementos del universo a cubrir. |
| **Columnas** | $J = \{1, 2, \dots, n\}$ | Conjunto de columnas (conjuntos candidatos). |
| **Matriz de Cobertura** | $A \in \{0, 1\}^{m \times n}$ | Matriz binaria con coeficientes $a_{ij}$ que indican si la columna $j$ cubre la fila $i$. |
| **Subconjunto de Cobertura** | $S_i \subseteq J$ | Subconjunto de columnas que cubren la fila $i$, definido por: <br> $$S_i = \{j \in J \mid a_{ij} = 1\}$$ |



### Parámetros del Modelo

| Parámetro | Notación | Descripción |
| :--- | :--- | :--- |
| **Número de filas** | $m = 50$ | Cantidad de restricciones de cobertura. |
| **Número de columnas** | $n = 200$ | Cantidad de variables de decisión. |
| **Costo por columna** | $c_j \in \mathbb{Z}^+$ | Costo asociado a seleccionar la columna $j$, para todo $j \in J$. |
| **Matriz de incidencia** | $a_{ij} \in \{0, 1\}$ | Valor binario: $1$ si la columna $j$ cubre la fila $i$, $0$ en caso contrario. |


### Variable de Decisión

| Variable | Notación | Descripción |
| :--- | :--- | :--- |
| **Selección de columna** | $x_j \in \{0, 1\}$ | Variable binaria para todo $j \in J$ que determina la inclusión de la columna en la solución. |

La variable se define formalmente como:

$$x_j = \begin{cases} 1 & \text{si la columna } j \text{ es seleccionada en la solución} \\ 0 & \text{si la columna } j \text{ NO es seleccionada} \end{cases}$$


### Modelo de Programación Entera Binaria (IP)

El modelo tiene como objetivo minimizar el costo total de las columnas seleccionadas, asegurando que cada fila sea cubierta por al menos una columna.

**Formulación:**

$$\min \quad z = \sum_{j \in J} c_j \cdot x_j \quad \quad \quad \quad \quad \text{(FO)}$$

**Sujeto a:**

$$\sum_{j \in S_i} x_j \geq 1, \quad \forall i \in I \quad \quad \quad \quad \quad \quad \quad \quad \text{(Cobertura)}$$

$$x_j \in \{0, 1\}, \quad \forall j \in J \quad \quad \quad \quad \quad \quad \quad \quad \text{(Binariedad)}$$


### Interpretación del Modelo

El problema de cobertura de conjuntos (*Set Covering Problem*) se puede interpretar desde una perspectiva operativa y de optimización de la siguiente manera:

* **Requerimientos ($i \in I$):** Cada fila del modelo representa una necesidad o elemento del universo que debe ser cubierto obligatoriamente (por ejemplo: una zona geográfica que requiere cobertura de señal, un cliente por atender o una tarea por ejecutar).
* **Alternativas de Servicio ($j \in J$):** Cada columna representa una opción disponible para satisfacer dichos requerimientos (por ejemplo: la ubicación de una antena, una ruta de distribución o la asignación de un recurso). Cada alternativa tiene un costo asociado $c_j$ y cubre un subconjunto específico de requerimientos $S_i$.

El modelo busca seleccionar el subconjunto óptimo de alternativas de servicio que garantice la atención de todos los requerimientos al menor costo posible.

## 2. Modelo de Programación Entera Binaria (IP)

La formulación matemática del modelo es la siguiente:

$$\min \quad z = \sum_{j \in J} c_j x_j$$

$$\text{s.t.} \quad \sum_{j \in S_i} x_j \geq 1 \qquad \forall \, i \in I$$

$$x_j \in \{0, 1\} \qquad \forall \, j \in J$$

Donde $x_j = 1$ si se selecciona la columna $j$ y $x_j = 0$ en caso contrario, $S_i$ es el conjunto de columnas que cubren la fila $i$, y $c_j$ es el costo de seleccionar la columna $j$.

El modelo se resuelve con *Branch & Bound*.

```python

import gurobipy as gp
from gurobipy import GRB
import time

t0 = time.time()

model = gp.Model("SCP_50x200")
model.setParam("OutputFlag", 1)    # mostrar log del solver
model.setParam("TimeLimit",  600)  # maximo 10 minutos
model.setParam("MIPGap",     1e-5) # detener si gap < 0.001%
model.setParam("Presolve",   2)    # presolve agresivo
model.setParam("MIPFocus",   1)    # enfocarse en encontrar buenas soluciones enteras
model.setParam("Cuts",       2)    # agregar cortes agresivamente

# Variables de decisión
x = model.addVars(N, vtype=GRB.BINARY, name="x")

# Función objetivo: minimizar costo total
model.setObjective(gp.quicksum(c[j] * x[j] for j in range(N)), GRB.MINIMIZE)

# Restricciones de cobertura: cada fila debe ser cubierta por al menos una columna
for i in range(M):
    model.addConstr(
        gp.quicksum(x[j] for j in S[i]) >= 1,
        name=f"cover_{i+1}"
    )

model.optimize()

t_ip = time.time() - t0

print(f"\nResolucion IP completada en {t_ip:.2f} segundos")
```

```python

import gurobipy as gp
from gurobipy import GRB


z_IP    = model.ObjVal
z_bound = model.ObjBound
gap     = model.MIPGap * 100
n_nodos = int(model.NodeCount)

seleccionadas = sorted([j for j in range(N) if x[j].X > 0.5])
n_sel = len(seleccionadas)
costo_real = sum(c[j] for j in seleccionadas)


print("SOLUCIÓN OPTIMA - MODELO IP")

print(f"  Costo optimo (z*)          : {z_IP:.0f}")
print(f"  Costo verificado           : {costo_real}  (debe coincidir)")
print(f"  Cota inferior del solver   : {z_bound:.4f}")
print(f"  Gap de optimalidad         : {gap:.6f}%")
print(f"  Tiempo de resolucion       : {t_ip:.2f} s")
print(f"  Nodos explorados (B&B)     : {n_nodos:,}")
print(f"  Columnas seleccionadas     : {n_sel} de {N}")
print()

# Verificacion de factibilidad
sin_cubrir = [i+1 for i in range(M) if not any(j in seleccionadas for j in S[i])]
if sin_cubrir:
    print(f"  ADVERTENCIA: filas sin cubrir -> {sin_cubrir}")
else:
    print(f"  Verificacion OK: las {M} filas estan cubiertas.")

print()
print(f"  Columnas seleccionadas (indices base 1):")
print(f"  {[j+1 for j in seleccionadas]}")

print()
print("  Detalle de cada columna seleccionada:")
print(f"  {'Col':>7}  {'Costo':>6}  {'Filas que cubre en el universo'}")

for j in seleccionadas:
    filas = [i+1 for i in filas_de[j]]
    print(f"  x_{j+1:<5}  {c[j]:>6}  {filas}")

print()
print("  Cobertura por fila (cuantas columnas seleccionadas cubren cada fila):")
print(f"  {'Fila':>5}  {'# cols':>6}  Columnas que la cubren")

redundantes = 0
for i in range(M):
    cubren = [j+1 for j in S[i] if j in seleccionadas]
    nota = " <- cubierta mas de una vez" if len(cubren) > 1 else ""
    if len(cubren) > 1:
        redundantes += 1
    print(f"  {i+1:>5}  {len(cubren):>6}  {cubren}{nota}")

print(f"\n  Filas con cobertura redundante: {redundantes} de {M}")
print("  (Normal en SCP: ocurre cuando columnas baratas comparten filas)")
```

## 4. Relajación LP Continua

Para realizar el análisis de sensibilidad es necesario resolver la relajación continua del modelo: se reemplaza $x_j \in \{0,1\}$ por $0 \leq x_j \leq 1$. Esto proporciona tres cosas:

1. Una cota inferior $z_{LP} \leq z_{IP}$ (*lower bound*).
2. Las variables duales $\pi_i$ (precios sombra) de cada restricción de cobertura.
3. Los costos reducidos $\bar{c}_j$ de cada variable.

Se usa el Simplex Dual (`Method=1`) porque garantiza duales exactos, y se desactiva el presolve para preservar la estructura original de las restricciones.

```python
lp = gp.Model("SCP_LP_relajado")
lp.setParam("OutputFlag", 0)
lp.setParam("Method", 1)    # simplex dual -> duales exactos
lp.setParam("Presolve", 0)  # sin presolve: preservar estructura de las restricciones

x_lp = lp.addVars(N, lb=0.0, ub=1.0, vtype=GRB.CONTINUOUS, name="x")

lp.setObjective(gp.quicksum(c[j] * x_lp[j] for j in range(N)), GRB.MINIMIZE)

restricciones_lp = []
for i in range(M):
    r = lp.addConstr(
        gp.quicksum(x_lp[j] for j in S[i]) >= 1,
        name=f"cover_{i+1}"
    )
    restricciones_lp.append(r)

t_lp_start = time.time()
lp.optimize()
t_lp = time.time() - t_lp_start

z_LP = lp.ObjVal
gap_integrality = z_IP - z_LP
gap_int_pct = gap_integrality / z_LP * 100


print("RELAJACIÓN LP - COTA INFERIOR")
print(f"  z_LP (relajación continua)  : {z_LP:.6f}")
print(f"  z_IP (solución entera)      : {z_IP:.0f}")
print(f"  Brecha de integralidad      : {gap_integrality:.6f}  ({gap_int_pct:.4f}%)")
print(f"  Tiempo LP                   : {t_lp:.4f} s")

# Clasificacion de variables LP
xv = [x_lp[j].X for j in range(N)]
en_1   = [j for j in range(N) if xv[j] > 1 - 1e-8]
en_0   = [j for j in range(N) if xv[j] < 1e-8]
frac   = [j for j in range(N) if 1e-8 <= xv[j] <= 1 - 1e-8]

print()
print(f"  Variables con x_j = 1        : {len(en_1)}")
print(f"  Variables con x_j = 0        : {len(en_0)}")
print(f"  Variables fraccionarias       : {len(frac)}")
if frac:
    print(f"  -> Las variables fraccionarias son las que obligan al B&B")
    print(f"     a ramificar para recuperar la integralidad.")
    for j in frac:
        print(f"     x_{j+1}: valor LP = {xv[j]:.5f}, costo = {c[j]}, filas que cubre: {[i+1 for i in filas_de[j]]}")

print()
brecha_str = "MUY BAJA -> la relajacion LP es una cota excelente del IP" if gap_int_pct < 2 else \
             "MODERADA -> el B&B necesita trabajo para cerrar la brecha"
print(f"  Interpretacion brecha ({gap_int_pct:.2f}%): {brecha_str}")
print(f"  Esto es consistente con lo reportado en Caprara et al. (1999)")
print(f"  para instancias densas del OR-Library.")
```

 ## 5. Análisis de sensibilidad: precios sombra ($\pi_i$)

El precio sombra $\pi_i$ de la restricción `cover_i` responde a la siguiente pregunta: si se exige que la fila $i$ sea cubierta por al menos $1 + \delta$ columnas en lugar de $1$, ¿cuánto aumenta el costo óptimo del LP? La respuesta es que $z_{LP}$ aumenta en $\pi_i \cdot \delta$, siempre que el cambio esté dentro del rango de sensibilidad válido.

- $\pi_i > 0$: restricción **activa** (*binding*). La fila $i$ es un cuello de botella: es costosa de cubrir marginalmente.
- $\pi_i = 0$: restricción **holgada**. La fila ya queda cubierta de sobra en la solución LP, por lo que relajarla no cambia el costo.

Identificar las filas con mayor $\pi_i$ permite priorizar dónde enfocar recursos si se quisieran cambiar los requisitos de cobertura del problema.

```python

pi     = [restricciones_lp[i].Pi    for i in range(M)]
holgura = [restricciones_lp[i].Slack for i in range(M)]

activas  = [i+1 for i in range(M) if abs(holgura[i]) < 1e-6]
holgadas = [i+1 for i in range(M) if abs(holgura[i]) >= 1e-6]

print("=" * 60)
print("ANALISIS DE SENSIBILIDAD - PRECIOS SOMBRA (pi_i)")
print("=" * 60)
print(f"  Restricciones activas (pi > 0)  : {len(activas)}  -> filas {activas}")
print(f"  Restricciones holgadas (pi = 0) : {len(holgadas)}  -> filas {holgadas}")
print()
print(f"  {'Fila':>5}  {'pi_i':>10}  {'Holgura':>10}  Estado")
print(f"  {'-'*5}  {'-'*10}  {'-'*10}  {'-'*35}")
for i in range(M):
    estado = "ACTIVA - cuello de botella" if abs(holgura[i]) < 1e-6 else "holgada - cubierta de sobra"
    print(f"  {i+1:>5}  {pi[i]:>10.4f}  {holgura[i]:>10.4f}  {estado}")

print()
print("  Top 5 filas con mayor precio sombra (las mas costosas de cubrir):")
top5 = sorted(range(M), key=lambda i: -pi[i])[:5]
print(f"  {'Fila':>5}  {'pi_i':>10}  {'Opciones de cobertura':>22}  Columnas disponibles")
print(f"  {'-'*5}  {'-'*10}  {'-'*22}  {'-'*35}")
for i in top5:
    print(f"  {i+1:>5}  {pi[i]:>10.4f}  {len(S[i]):>22}  {[j+1 for j in S[i]]}")

print()
print("  Lectura: una fila con pi_i alto y pocas opciones de cobertura")
print("  es el mayor cuello de botella del problema.")
```

## 6. Análisis de sensibilidad: costos reducidos ($\bar{c}_j$)

El costo reducido $\bar{c}_j$ mide qué tan lejos está la columna $j$ de ser atractiva para entrar o salir de la solución óptima del LP:

$$\bar{c}_j = c_j - \sum_{\substack{i \in I \\ j \in S_i}} \pi_i$$

- $\bar{c}_j > 0$ y $x_j = 0$: la columna es **cara**. Para que entrara a la solución, su costo tendría que bajar al menos $\bar{c}_j$ unidades.
- $\bar{c}_j < 0$ y $x_j = 1$: la columna es **rentable**. Su valor dual supera su costo. Incluso si $c_j$ subiera en $|\bar{c}_j|$, la columna seguiría en la solución óptima.
- $\bar{c}_j = 0$: la columna está en la **frontera** de indiferencia.

```python
rc = [x_lp[j].RC for j in range(N)]

# Verificacion manual del costo reducido
rc_manual = [c[j] - sum(pi[i] for i in filas_de[j]) for j in range(N)]

print("=" * 60)
print("ANALISIS DE SENSIBILIDAD - COSTOS REDUCIDOS")
print("=" * 60)
print(f"  {'Col':>7}  {'c_j':>5}  {'x_j LP':>8}  {'c_barra_j':>11}  {'c_barra manual':>14}  Estado")
print(f"  {'-'*7}  {'-'*5}  {'-'*8}  {'-'*11}  {'-'*14}  {'-'*22}")
for j in range(N):
    if xv[j] > 1 - 1e-8:
        estado = "seleccionada (x=1)"
    elif xv[j] < 1e-8:
        estado = "excluida (x=0)"
    else:
        estado = f"fraccionaria ({xv[j]:.3f})"
    print(f"  x_{j+1:<5}  {c[j]:>5}  {xv[j]:>8.5f}  {rc[j]:>11.4f}  {rc_manual[j]:>14.4f}  {estado}")

# Las mas rentables (en solucion, costo reducido mas negativo)
en_sol = [(j, rc[j]) for j in range(N) if xv[j] > 1 - 1e-8]
fuera  = [(j, rc[j]) for j in range(N) if xv[j] < 1e-8]

print()
print("  Top 5 columnas MAS RENTABLES en la solucion LP (c_barra mas negativo):")
print("  -> Son las columnas que mas 'vale la pena' tener. Podrian subir de")
print("     precio sin salir de la solucion optima.")
print(f"  {'Col':>7}  {'c_j':>5}  {'c_barra_j':>11}  Puede subir c_j hasta")
print(f"  {'-'*7}  {'-'*5}  {'-'*11}  {'-'*25}")
for j, rc_j in sorted(en_sol, key=lambda t: t[1])[:5]:
    tope = c[j] - rc_j
    print(f"  x_{j+1:<5}  {c[j]:>5}  {rc_j:>11.4f}  {tope:.2f}")

print()
print("  Top 5 columnas EXCLUIDAS mas cercanas a entrar (c_barra mas pequeno):")
print("  -> Si su costo bajara en c_barra_j unidades, entrarian a la solucion.")
print(f"  {'Col':>7}  {'c_j':>5}  {'c_barra_j':>11}  Reduccion necesaria en c_j")
print(f"  {'-'*7}  {'-'*5}  {'-'*11}  {'-'*30}")
for j, rc_j in sorted(fuera, key=lambda t: t[1])[:5]:
    print(f"  x_{j+1:<5}  {c[j]:>5}  {rc_j:>11.4f}  bajar al menos {rc_j:.2f}")

```

## 7. Análisis de Sensibilidad: ranging de costos

El ranging de costos responde a: ¿cuánto puede variar $c_j$ sin que la **base óptima** del LP cambie? Para cada columna $j$ se reporta el intervalo $[c_j^{\text{low}},\ c_j^{\text{up}}]$:

- Si $c_j$ baja de $c_j^{\text{low}}$: otra columna querría entrar a la base.
- Si $c_j$ sube de $c_j^{\text{up}}$: la columna $j$ querría salir de la base.

Un rango amplio indica una solución **robusta** ante variaciones en ese costo. Un rango estrecho indica una solución **sensible**: pequeños cambios en $c_j$ alteran la composición de la solución óptima.

```python

print("ANALISIS DE SENSIBILIDAD - RANGING DE COSTOS")

print(f"  {'Col':>7}  {'c_j':>5}  {'c_j_low':>11}  {'c_j_up':>11}  {'Amplitud':>10}  Robustez")


ranging_data = []
for j in range(N):
    lo  = x_lp[j].SAObjLow
    hi  = x_lp[j].SAObjUp
    amp = (hi - lo) if (hi < 1e29 and lo > -1e29) else float("inf")
    ranging_data.append((j, c[j], lo, hi, amp))
    lo_s  = f"{lo:9.4f}" if lo  > -1e29 else "       -inf"
    hi_s  = f"{hi:9.4f}" if hi  <  1e29 else "       +inf"
    amp_s = f"{amp:8.4f}" if amp <  1e29 else "         inf"
    rob   = "alta" if amp > 50 else ("media" if amp > 10 else "baja")
    print(f"  x_{j+1:<5}  {c[j]:>5}  {lo_s:>11}  {hi_s:>11}  {amp_s:>10}  {rob}")

finitos = [(j,cj,lo,hi,amp) for j,cj,lo,hi,amp in ranging_data if amp < 1e29]

if finitos:
    print()
    print("  Top 5 columnas con rango MAS ESTRECHO (mayor sensibilidad):")
    print(f"  {'Col':>7}  {'c_j':>5}  {'c_j_low':>9}  {'c_j_up':>9}  {'Amplitud':>10}")

    for j,cj,lo,hi,amp in sorted(finitos, key=lambda t: t[4])[:5]:
        print(f"  x_{j+1:<5}  {cj:>5}  {lo:>9.4f}  {hi:>9.4f}  {amp:>10.4f}")

    print()
    print("  Top 5 columnas con rango MAS AMPLIO (mayor robustez):")
    print(f"  {'Col':>7}  {'c_j':>5}  {'c_j_low':>9}  {'c_j_up':>9}  {'Amplitud':>10}")

    for j,cj,lo,hi,amp in sorted(finitos, key=lambda t: -t[4])[:5]:
        lo_s = f"{lo:.4f}" if lo > -1e29 else "-inf"
        hi_s = f"{hi:.4f}" if hi <  1e29 else "+inf"
        print(f"  x_{j+1:<5}  {cj:>5}  {lo_s:>9}  {hi_s:>9}  {amp:>10.4f}")
```

## 8. Análisis de Sensibilidad: ranging del RHS

Actualmente cada restricción `cover_i` tiene lado derecho $b_i = 1$, lo que significa que la fila $i$ debe ser cubierta por al menos una columna. El ranging del RHS $[b_i^{\text{low}},\ b_i^{\text{up}}]$ indica dentro de qué rango puede moverse $b_i$ sin que la base óptima cambie.

Dentro del rango, el cambio en el costo óptimo es exactamente:

$$\Delta z_{LP} = \pi_i \cdot \Delta b_i$$

Por ejemplo, si se exige que la fila $i$ esté cubierta por al menos 2 columnas ($b_i = 2$), el costo sube en $\pi_i \cdot 1$, siempre que $2 \leq b_i^{\text{up}}$.

```python

print("ANALISIS DE SENSIBILIDAD - RANGING DEL RHS")

print(f"  {'Fila':>5}  {'b_i':>4}  {'b_i_low':>11}  {'b_i_up':>11}  {'pi_i':>9}  Efecto si b_i aumenta en 1")

for i in range(M):
    lo = restricciones_lp[i].SARHSLow
    hi = restricciones_lp[i].SARHSUp
    lo_s = f"{lo:9.4f}" if lo > -1e29 else "       -inf"
    hi_s = f"{hi:9.4f}" if hi <  1e29 else "       +inf"
    efecto = f"z_LP sube {pi[i]:.4f}" if pi[i] > 1e-8 else "z_LP no cambia (restriccion holgada)"
    print(f"  {i+1:>5}  {'1':>4}  {lo_s:>11}  {hi_s:>11}  {pi[i]:>9.4f}  {efecto}")
```

## Verificación de condiciones KKT

Las condiciones de Karush-Kuhn-Tucker (KKT) son las condiciones necesarias y suficientes de optimalidad para el LP. Verificarlas confirma que la solución obtenida es genuinamente óptima. Las cuatro condiciones son:

1. **Factibilidad primal:** $\displaystyle\sum_{j \in S_i} x_j \geq 1 \quad \forall\, i \in I$

2. **Factibilidad dual:** $\pi_i \geq 0 \quad \forall\, i \in I$

3. **Holgura complementaria primal-dual:** $\displaystyle\pi_i \cdot \left(\sum_{j \in S_i} x_j - 1\right) = 0 \quad \forall\, i \in I$ — si $\pi_i > 0$ entonces la restricción es activa, y viceversa.

4. **Holgura complementaria dual-primal:** $\bar{c}_j \cdot x_j = 0 \quad \forall\, j \in J$ — si $x_j > 0$ entonces $\bar{c}_j = 0$, y viceversa.


```python
print("VERIFICACIÓN DE CONDICIONES KKT")

ok_count = 0

# (1) Factibilidad primal - garantizada por el solver, solo verificamos
fact_prim = all(sum(xv[j] for j in S[i]) >= 1 - 1e-6 for i in range(M))
estado = "OK" if fact_prim else "NO CUMPLE"
print(f"  (1) Factibilidad primal (Ax >= b)          : {estado}")
if fact_prim: ok_count += 1

# (2) Factibilidad dual
errores_dual = [i+1 for i in range(M) if pi[i] < -1e-6]
estado = "OK" if not errores_dual else f"NO CUMPLE en filas {errores_dual}"
print(f"  (2) Factibilidad dual (pi_i >= 0)          : {estado}")
if not errores_dual: ok_count += 1

# (3) Holgura complementaria primal-dual
errores_cp = []
for i in range(M):
    cobertura_i = sum(xv[j] for j in S[i])
    if pi[i] > 1e-6 and abs(cobertura_i - 1.0) > 1e-5:
        errores_cp.append(i+1)

estado = "OK" if not errores_cp else f"NO CUMPLE en filas {errores_cp}"
print(f"  (3) Holgura complementaria primal-dual     : {estado}")
if not errores_cp: ok_count += 1

# (4) Holgura complementaria dual-primal
errores_cd = [
    j+1
    for j in range(N)
    if xv[j] > 1e-6 and abs(rc[j]) > 1e-4
]

estado = "OK" if not errores_cd else f"NO CUMPLE en {len(errores_cd)} columnas"
print(f"  (4) Holgura complementaria dual-primal     : {estado}")
if not errores_cd: ok_count += 1

print()

if ok_count == 4:
    print("  Todas las condiciones KKT se satisfacen.")
    print("  La solución LP obtenida es óptima.")
else:
    print(f"  Atención: {4 - ok_count} condición(es) KKT requieren revisión.")
```
